# 1. Fill Null And Missing Values

```python
# Create a random Series of integers between 18 and 60 in the age column.

rand_vals = pd.Series(
    np.random.randint(18, 60, size=og['Age'].isna().sum()),
    index=og.index[og['Age'].isna()]            # align indices
)
og['Age'] = og['Age'].fillna(rand_vals)
```

```python
# Create a random Series of integers between 110 and 220 in the Height column.

rand_vals = pd.Series(
    np.random.randint(110, 220, size=og['Height'].isna().sum()),
    index=og.index[og['Height'].isna()]            # align indices
)
og['Height'] = og['Height'].fillna(rand_vals)
```

```python
# Create a random Series of integers betweenv60 and 240 in the Weight column.

rand_vals = pd.Series(
    np.random.randint(60, 240, size=og['Weight'].isna().sum()),
    index=og.index[og['Weight'].isna()]            # align indices
)
og['Weight'] = og['Weight'].fillna(rand_vals)
```

## 16. Count unique athletes per year and display plot.

```python
athletes_per_year = og.groupby('Year')['ID'].nunique()
athletes_per_year.plot(figsize=(10,6), title='Unique Athletes per Olympic Year')
plt.xlabel('Year')
plt.ylabel('Number of Athletes')
plt.show()
```

## 17. Count medals per country and get top 20

```python
medals = og.dropna(subset=['Medal'])
top_countries = medals.groupby('NOC')['Medal'].count().sort_values(ascending=False).head(20)
top_countries.sort_values().plot(kind='barh', figsize=(10,6), title='Top 20 Countries by Total Medals')
plt.xlabel('Total Medals')
plt.show()
```

## 18. Show data that Male vs Female Participation Over Time using plot

## 19. Show data of Age Distribution of Medalists using plot.

## 20. Height vs Weight - Basketball Athletes

```python
basketball = og[og['Sport']=='Basketball']
colors = basketball['Medal'].map({'Gold':'gold','Silver':'silver','Bronze':'brown'}).fillna('blue')
plt.figure(figsize=(10,6))
plt.scatter(basketball['Height'], basketball['Weight'], c=colors)
plt.title('Height vs Weight - Basketball Athletes')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
